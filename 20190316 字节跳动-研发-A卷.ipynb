## 题目1 找硬币

目前晓明 有 1024 元钱的纸币，他想买一个 N 元的物品，已知这个世界有 64,16,4,1 币值 的硬币，请问最少找多少枚硬币？

输入:
> 物品价格 N 

输出：
> 最少找回硬币数量

举例:

输入:
> 200

输出：
> 17

```python
N = int(input().strip())
All = 1024
Cone = [64,16,4,1]
re_back = All-N
cone_list=[]
for cone in Cone:
    count = re_back//cone
    re_back-= count*cone
    cone_list.append(count)
print(sum(cone_list))
```

## 题目2 王大锤自动校正拼写错误

王大锤 发现校对工作是很麻烦的工作，总是会出现拼写错误：

1. `AAA` 形式错误 是由于3个重复造成的，去掉一个则正确，例如 `helllo` -> `hello`;
2. `AABB` 形式错误 需要去掉后面重复的，例如 `helloo` -> `hello`;
3. 匹配按照从左到右的顺序，例如 `AABBCC` -> `AABCC`;

输入:
> n 个测试用例  
> 字符串 
> ... 共n个

输出：
> 校对过的 字符串   
> 与输入对应个数

举例:

输入:
> 2  
> helloo  
> wooooooooooooow

输出：
> hello  
> woow

```python
def mask(input_str):
    lenth= len(input_str)
    idx = 0
    while lenth > 2:
        if lenth > 3 and input_str[idx] == input_str[idx+1] and input_str[idx+2] == input_str[idx+3]:
            input_str = input_str[:idx+2]+input_str[idx+3:]
        elif input_str[idx] == input_str[idx+1] == input_str[idx+2]:
            input_str = input_str[:idx+1]+input_str[idx+2:]
        else:
            idx+=1
        lenth= len(input_str[idx:])
        # print(input_str,idx,lenth)
    print(input_str)
        
n = int(input().strip())
words_list = []
for _ in range(n):
    words_list.append(input().strip())
for word in words_list:
    mask(word)
```

## 题目3 发奖品

小朋友们 测试，每个人过后都会得到一个分数，老师让他们站成一圈，每个人都会至少得到一块糖果，如果小朋友的分数比相邻他左右的小朋友都高，那么他得到的糖数也比左右的小朋友多。给定 小朋友数，然后给出他们对应的分数，求需要准备多少糖果？


输入:
> n 个测试用例  
> 小朋友数 
> 每个小朋友的数量  
>  
> 第n组用例

输出：
> 准备的糖果数   
> 与输入测试用例对应糖果数 输出

举例:

输入:
> 2  
> 2  
> 1 2  
> 4  
> 1 2 3 3

输出：
> 3   
> 8  

提示：  

1 2 3 3 分数 按照规则 分发的糖果数依次为 1 2 3 2

```python
n = int(input().strip())
task_list = []
for _ in range(n):
    peopl_num= int(input().strip())
    score_list = [int(num) for num in input().strip().split(' ')]
    task_list.append([peopl_num,score_list])
for p_num,scor_list in task_list:
    gift_list = [1 for _ in scor_list] # 初始化全1
    length = len(scor_list)
    for idx,scro in enumerate(scor_list):
        left,right= idx-1,idx+1 if idx<(length-1) else 0
        if scro > scor_list[left]:
            gift_list[idx] = gift_list[left]+1
        if scro > scor_list[right]:
            gift_list[idx] = gift_list[right]+1
    # 二次检查
    for idx,scro in enumerate(scor_list):
        idx = (-idx)%length
        scro = scor_list[idx]
        left,right= idx-1,idx+1 if idx<(length-1) else 0
        if scro > scor_list[left]:
            gift_list[idx] = max((gift_list[left]+1),gift_list[idx])
        if scro > scor_list[right]:
            gift_list[idx] = max((gift_list[right]+1),gift_list[idx])
        # print(left,idx,right,gift_list)
    print(sum(gift_list))
```

## 题目4 求截取等长绳子的最大长度

目前有 N 段绳子，我们需要 M 段等长的绳子，需要从N段绳子中截取(只截取，不连接)，求 M 段绳子 最大长多是多少？

输入:
> 绳子个数 N 需要的绳子段数 M  
> 每段已有绳子的长度

输出：
> 最大截取长度 保留小数两位

举例:

输入:
> 3 4  
> 3 5 4

输出：
> 2.50

提示:

绳子截取为 2.50 2.50 2.50 2.50 ,其中长度为 5 的绳子正好截取两段

```python
N,M = [int(num) for num in input().strip().split(' ')]
L_list = [int(num) for num in input().strip().split(' ')]
length = min(L_list)
while True:
    line_count=[]
    for line in L_list:
        line_count.append(line//length)  
    if sum(line_count) >= M:
        break
    else:
        length = float('%.2f'%(length-0.01))
print('%.2f'%length)
```
